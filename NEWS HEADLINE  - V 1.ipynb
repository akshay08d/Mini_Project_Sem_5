```python
test_dset2 = test_dset1.drop(columns=['tweet_id','airline_sentiment_confidence','negativereason','negativereason_confidence','airline','airline_sentiment_gold','name','negativereason_gold','retweet_count','tweet_coord','tweet_created','tweet_location','user_timezone'])
```

```python
test_dset2.head(5)
```

```python
test_dset2.drop(test_dset2[test_dset2['airline_sentiment'] == 'neutral'].index, inplace = True)
```

```python
test_dset2.airline_sentiment.value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=1100)
vectors = vectorizer.fit_transform(test_dset2.text)
news_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names())
news_df.head()
```

```python
X = news_df
y = test_dset2.airline_sentiment
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
```

```python
%%time
# Create and train a logistic regression
logreg = LogisticRegression(C=1e9, solver='lbfgs', max_iter=2000)
logreg.fit(X, y)
```

```python
%%time
# Create and train a random forest classifier
forest = RandomForestClassifier(n_estimators=50)
forest.fit(X, y)
```

```python
%%time
# Create and train a linear support vector classifier (LinearSVC)
svc = LinearSVC()
svc.fit(X, y)
```

```python
%%time
# Create and train a multinomial naive bayes classifier (MultinomialNB)
bayes = MultinomialNB()
bayes.fit(X, y)
```
